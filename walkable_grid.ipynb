```python
import numpy as np

# A = [...]  # Your 36x30 matrix defined above

nodes = np.array(A, dtype=int)

# Check shape
assert nodes.shape == (36, 30), f"Matrix must be 36x30, got {nodes.shape}"

# Ensure only 0/1
nodes = (nodes != 0).astype(int)

NODES_Y = 36
NODES_X = 30
DEPTH   = NODES_Y * NODES_X   # 1080
WIDTH   = 1                   # <-- 1 BIT WIDE ROM

mif_lines = []
mif_lines.append(f"DEPTH = {DEPTH};")
mif_lines.append(f"WIDTH = {WIDTH};")
mif_lines.append("ADDRESS_RADIX = HEX;")
mif_lines.append("DATA_RADIX = BIN;")   # <-- BINARY OUTPUT
mif_lines.append("")
mif_lines.append("CONTENT BEGIN")

# Write entries: row-major order (ny, nx)
for ny in range(NODES_Y):
    for nx in range(NODES_X):
        addr = ny * NODES_X + nx
        bit  = nodes[ny, nx]

        addr_str = f"{addr:03X}"   # hex address 000..437
        data_str = f"{bit:d}"      # '0' or '1'

        mif_lines.append(f"    {addr_str} : {data_str};")

mif_lines.append("END;")

mif_text = "\n".join(mif_lines)

with open("node.mif", "w") as f:
    f.write(mif_text)

print("Created node.mif (WIDTH=1, DEPTH=1080)")

from google.colab import files
files.download("node.mif")
```
